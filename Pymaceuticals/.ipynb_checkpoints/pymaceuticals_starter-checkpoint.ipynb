### Analysis

- Observation One: The correlation between mouse size and tumor size is .84. This is a strong positive correlation meaning that when the mouse size increases so does the tumor size. 

- Obervation Two: Capumolin seems to most effectively reduce tumor size of the treatments tested on these mice. 

- Observation Three: Gender did not impact efficacy of treatment
 

```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse = pd.read_csv(mouse_metadata_path)
study = pd.read_csv(study_results_path)

# Combine the dataframes
df = pd.merge(mouse, study, how="left", on="Mouse ID")

# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
len(df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_ids = df.loc[df.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data = df.loc[df["Mouse ID"]=="g989"]
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = df[df["Mouse ID"].isin(duplicate_ids)==False]
clean_df.head(25)
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

means = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
medians = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variances = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
sds = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary_table = pd.DataFrame({"Mean Tumor Volume":means,
                                "Median Tumor Volume":medians,
                                "Tumor Volume Variance":variances,
                                "Tumor Volume Std. Dev.":sds,
                                "Tumor Volume Std. Err.":sems})
summary_table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summ_table = clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summ_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Get the last (greatest) timepoint for each mouse
max_timepoints = clean_df.groupby('Mouse ID')['Timepoint'].max()

# Create a DataFrame with 'Mouse ID' and the corresponding maximum 'Timepoint'
max_timepoints_df = pd.DataFrame(max_timepoints).reset_index()

# Merge the DataFrame with the greatest timepoint back to the original DataFrame
merged_df = pd.merge(clean_df, max_timepoints_df, on=['Mouse ID', 'Timepoint'], how='inner')

# Filter the merged DataFrame to include only the specified treatment regimens
specified_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
final_tumor_volumes = merged_df[merged_df['Drug Regimen'].isin(specified_regimens)]

# Display the DataFrame with the final tumor volume for each mouse across the specified treatment regimens
final_tumor_volumes
```

```python
# Create a list of treatment names and an empty list for tumor volume data
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_volumes = []

# Loop through each treatment and append final tumor volumes to the list
for treatment in treatments:
    final_tumor_volume = merged_df.loc[merged_df['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    tumor_volumes.append(final_tumor_volume)

# Calculate quartiles and IQR for each treatment regimen to identify outliers
for i, treatment in enumerate(treatments):
    Q1 = tumor_volumes[i].quantile(0.25)
    Q3 = tumor_volumes[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = tumor_volumes[i][(tumor_volumes[i] < lower_bound) | (tumor_volumes[i] > upper_bound)]
    
    if outliers.empty:
        print(f"No outliers for {treatment}")
    else:
        print(f"Potential outliers for {treatment}: {outliers}")
        
    # Additional analysis or visualization can be performed here
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

plt.figure(figsize=(10, 6))
plt.boxplot(tumor_volumes, labels=treatments, flierprops=dict(marker='o', markerfacecolor='red', markersize=8))
plt.title('Final Tumor Volume for Each Treatment Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xlabel('Treatment Regimen')
plt.grid(axis='y')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
cap_mice = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]
cap_mice.value_counts("Mouse ID")
Mouse_b128 = clean_df.loc[clean_df["Mouse ID"] == "b128"]


#Make variables to plot 
y = Mouse_b128["Tumor Volume (mm3)"]
x = Mouse_b128["Timepoint"]

# Plot our line that will be used to 
plt.plot(x, y, color="yellow", label="Tumor volume")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

#Give titles and labels to graph
plt.title( "Mouse b128s tumor volume over time")
plt.xlabel("Timepoint (Days)")
plt.ylabel(" Tumor volume (mm3)")

# Print our chart to the screen
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

cap_df = clean_df[clean_df['Drug Regimen'] == 'Capomulin']

mouse_weight_cap = cap_df.groupby(["Mouse ID"])["Weight (g)"].mean()

tumor_vol_cap = cap_df.groupby(["Mouse ID"])["Tumor Volume (mm3)"].mean()


#Make variables to plot 
x = mouse_weight_cap
y = tumor_vol_cap

# Plot our line that will be used to 
plt.scatter(x, y, color="purple", label="Tumor volume")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

#Give titles and labels to graph
plt.title( "Mouse Weight vs. Tumor Volume")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

# Print our chart to the screen
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

correlation, _ = st.pearsonr(mouse_weight_cap, tumor_vol_cap)
print(f"The correlation between both factors is: {correlation}")

# Perform linear regression to get the slope and intercept
slope, intercept = np.polyfit(mouse_weight_cap, tumor_vol_cap, 1)

# Create the regression line
regression_line = slope * mouse_weight_cap + intercept

# Plot the scatter plot of mouse weight vs. average tumor volume
plt.figure(figsize=(10, 6))
plt.scatter(mouse_weight_cap, tumor_vol_cap, color='b', label='Data Points')
plt.plot(mouse_weight_cap, regression_line, color='r', linestyle='-', label='Linear Regression')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.legend()
plt.grid(True)
plt.show()
```
